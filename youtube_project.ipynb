```python
import tkinter as tk
import webbrowser

# Define the URLs to open
url1 = "https://www.google.com"
url2 = "https://www.youtube.com"
url3 = "https://www.amazon.com"
url4 = "https://www.linkedin.com"

# Define the function to open a URL in a new window
def open_url(url):
    webbrowser.open_new(url)

# Create the main window and grid layout
root = tk.Tk()
root.geometry("500x500")
root.title("URL Grid Example")

# Create the grid layout using frames
frame1 = tk.Frame(root, bg="white")
frame1.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")
frame2 = tk.Frame(root, bg="white")
frame2.grid(row=0, column=1, padx=10, pady=10, sticky="nsew")
frame3 = tk.Frame(root, bg="white")
frame3.grid(row=1, column=0, padx=10, pady=10, sticky="nsew")
frame4 = tk.Frame(root, bg="white")
frame4.grid(row=1, column=1, padx=10, pady=10, sticky="nsew")

# Add buttons to each frame and bind them to open the URLs
button1 = tk.Button(frame1, text="Open URL 1", command=lambda: open_url(url1))
button1.pack(fill="both", expand=True)
button2 = tk.Button(frame2, text="Open URL 2", command=lambda: open_url(url2))
button2.pack(fill="both", expand=True)
button3 = tk.Button(frame3, text="Open URL 3", command=lambda: open_url(url3))
button3.pack(fill="both", expand=True)
button4 = tk.Button(frame4, text="Open URL 4", command=lambda: open_url(url4))
button4.pack(fill="both", expand=True)

# Start the main loop
root.mainloop()
```
